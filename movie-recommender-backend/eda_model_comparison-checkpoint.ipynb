```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from surprise import SVD, SVDpp, KNNBasic, KNNWithMeans, Dataset, Reader
from surprise.model_selection import cross_validate, GridSearchCV
```

```python
ratings = pd.read_csv("../ml-latest-small/ratings.csv")
movies = pd.read_csv("../ml-latest-small/movies.csv")
tags = pd.read_csv("../ml-latest-small/tags.csv")

ratings.head(), movies.head(), tags.head()
```

```python
plt.figure(figsize=(8, 4))
sns.countplot(x='rating', data=ratings)
plt.title('Ratings Distribution')
plt.show()
```

```python
ratings_per_user = ratings.groupby('userId').size()
plt.figure(figsize=(8, 4))
sns.histplot(ratings_per_user, bins=30, kde=False)
plt.title('Number of Ratings per User')
plt.xlabel('Number of Ratings')
plt.ylabel('Number of Users')
plt.show()
```

```python
most_rated = ratings['movieId'].value_counts().head(10)
top_movies = movies[movies['movieId'].isin(most_rated.index)]
top_movies['rating_count'] = top_movies['movieId'].map(most_rated)
top_movies[['title', 'rating_count']]
```

```python
movies['genres'] = movies['genres'].str.split('|')
genre_counts = pd.Series([g for sub in movies['genres'] for g in sub]).value_counts()

plt.figure(figsize=(10, 4))
genre_counts.plot(kind='bar', title='Genre Frequency')
plt.ylabel('Frequency')
plt.show()
```

```python
reader = Reader(rating_scale=(0.5, 5.0))
data = Dataset.load_from_df(ratings[['userId', 'movieId', 'rating']], reader)
```

```python
param_grid = {
    'n_epochs': [5, 10],
    'lr_all': [0.002, 0.005],
    'reg_all': [0.4, 0.6]
}
gs = GridSearchCV(SVD, param_grid, measures=['rmse', 'mae'], cv=3)
gs.fit(data)

print("Best RMSE (SVD):", gs.best_score['rmse'])
print("Best Parameters (SVD):", gs.best_params['rmse'])
```

```python
models = {
    'SVD': SVD(**gs.best_params['rmse']),
    'SVD++': SVDpp(),
    'KNNBasic': KNNBasic(),
    'KNNWithMeans': KNNWithMeans()
}

results = {}
for name, model in models.items():
    cv_results = cross_validate(model, data, measures=['RMSE', 'MAE'], cv=3, verbose=False)
    results[name] = {
        'RMSE': np.mean(cv_results['test_rmse']),
        'MAE': np.mean(cv_results['test_mae'])
    }

results_df = pd.DataFrame(results).T.reset_index().rename(columns={'index': 'Model'})
results_df
```

```python
plt.figure(figsize=(10, 5))
bar_width = 0.35
x = np.arange(len(results_df['Model']))

plt.bar(x - bar_width / 2, results_df['RMSE'], bar_width, label='RMSE', color='skyblue')
plt.bar(x + bar_width / 2, results_df['MAE'], bar_width, label='MAE', color='salmon')

plt.xlabel('Models')
plt.ylabel('Error')
plt.title('Model Comparison: RMSE vs MAE')
plt.xticks(x, results_df['Model'])
plt.legend()
plt.tight_layout()
plt.show()
```
